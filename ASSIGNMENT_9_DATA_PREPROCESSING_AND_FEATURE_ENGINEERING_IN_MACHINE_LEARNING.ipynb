```python
#MISSING VALUES
import numpy as np
df.replace('?', np.nan, inplace=True)
print(df.isnull().sum())
```

```python
# for doing the mode values
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

## Scaling Numerical Features

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(numerical_cols)
```

```python
#Standard Scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_Standard_Scaler=df.copy()
df_Standard_Scaler[numerical_cols]=scaler.fit_transform(df[numerical_cols])
```

## Encoding Techniques

```python
cat_cols = df.select_dtypes(include=['object']).columns

low_card = [col for col in cat_cols if df[col].nunique() < 5]
high_card = [col for col in cat_cols if df[col].nunique() >= 5]

print("Low Cardinality:", low_card)
print("High Cardinality:", high_card)
```

```python
#One-Hot Encoding (< 5 categories)
df = pd.get_dummies(df, columns=low_card, drop_first=True)
df
```

```python
#Label Encoding (≥ 5 categories)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in high_card:
    df[col] = le.fit_transform(df[col])
```

```python
df
```

Label Encoding

Advantages:
Memory efficient
Faster computation
Works well for Tree-based models (Random Forest, XGBoost)

Disadvantages:
Introduces false ordinal relationship
Example: "Doctor > Engineer > Teacher" (wrong assumption)
Can mislead linear models
